```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
print(X)
```

```python
print(y)
```

```python
# Standardize the feature values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print(X_scaled)
```

```python
# Instantiate PCA with 2 principal components
pca = PCA(n_components=2)     # comp can be take less than features
X_pca = pca.fit_transform(X_scaled)
```

```python
print(X_pca)
```

```python
# Instantiate LDA with 2 discriminant components
lda = LDA(n_components=2)      #min of no of feature or (no of class-1)
X_lda = lda.fit_transform(X_scaled, y)
```

```python
print(X_lda)
```

```python
# Plotting PCA results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_pca[y == 0, 0], X_pca[y == 0, 0], label='Setosa')
plt.scatter(X_pca[y == 1, 0], X_pca[y == 1, 0], label='Versicolor')
plt.scatter(X_pca[y == 2, 0], X_pca[y == 2, 0], label='Virginica')
plt.title('PCA of IRIS dataset')
plt.legend()
```

```python
# Plotting LDA results
plt.subplot(1, 2, 2)
plt.scatter(X_lda[y == 0, 0], X_lda[y == 0, 1], label='Setosa')
plt.scatter(X_lda[y == 1, 0], X_lda[y == 1, 1], label='Versicolor')
plt.scatter(X_lda[y == 2, 0], X_lda[y == 2, 1], label='Virginica')
plt.title('LDA of IRIS dataset')
plt.legend()
```

```python
plt.tight_layout()
plt.show()
```

```python
# Load the Wine dataset from a CSV file
wine_data = pd.read_csv('wine.csv')
X = wine_data.iloc[:, 1:]
y = wine_data.iloc[:, 0]
```

```python
# Standardize the feature values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
# LDA
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X_scaled, y)
```

```python
# Plotting PCA and LDA results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
for target_label in np.unique(y):
    plt.scatter(X_pca[y == target_label, 0], X_pca[y == target_label, 1], label=f'Class {target_label}')
plt.title('PCA of Wine Dataset')
plt.legend()

plt.subplot(1, 2, 2)
for target_label in np.unique(y):
    plt.scatter(X_lda[y == target_label, 0], X_lda[y == target_label, 1], label=f'Class {target_label}')
plt.title('LDA of Wine Dataset')
plt.legend()

plt.tight_layout()
plt.show()
```
